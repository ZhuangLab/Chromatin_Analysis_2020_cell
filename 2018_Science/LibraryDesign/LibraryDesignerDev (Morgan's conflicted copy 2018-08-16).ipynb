```python
import numpy as np
word=16
base = 4**(np.arange(word,dtype='uint64')[::-1])
def seq2Int3(seq):
    #dic_conv = {'C':1,'c':1,'G':2,'g':2,'T':3,'t':3}
    #digits = np.array([dic_conv.get(let,0) for let in seq],dtype='uint64')
    word = len(seq)
    digits = np.zeros(word,dtype='uint64')
    count = 0
    for let in seq:
        if let=='C':
            digits[count]=1
        elif let=='c':
            digits[count]=1
        elif let=='G':
            digits[count]=2
        elif let=='g':
            digits[count]=2
        elif let=='T':
            digits[count]=3
        elif let=='t':
            digits[count]=3
        count+=1
    #base = 4**(np.arange(word,dtype='uint64')[::-1])
    return np.sum(base*digits)
```

```python
def fastaread(fl,force_upper=False):
    fid = open(fl,'r')
    names = []
    seqs = []
    lines = []
    while True:
        line = fid.readline()
        if not line:
            seq = "".join(lines)
            if force_upper:
                seq=seq.upper()
            seqs.append(seq)
            break
        if line[0]=='>':
            name = line[1:-1]
            names.append(name)
            seq = "".join(lines)
            if force_upper:
                seq=seq.upper()
            seqs.append(seq)
            lines = []
        else:
            lines.append(line[:-1])
    fid.close()
    return [names,seqs[1:]]
```
